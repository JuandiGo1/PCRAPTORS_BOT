```python
import matplotlib.pyplot as plt
```

```python
class star:
    def __init__(self, x, y, id,HRN, name=None) :
        self.x = x
        self.y = y
        self.id=id
        self.HRN= HRN #Harvard Revised Number
        self.name = name
```

```python

stars=[]
with open('stars.txt', 'r') as archivo:
    # Leer cada línea del archivo y hacer split por espacio
    for linea in archivo:
        datos = linea.split()
        nombres= []
        if len(datos) >6:
            ind= 6
            cad=""
            while ind < len(datos):
                nombres.append(datos[ind])
                ind+=1
            for i in nombres:
                cad+=i

            nombres = cad.split(";")
        Star = star(float(datos[0]), float(datos[1]), float(datos[3]),float(datos[5]), nombres)
        stars.append(Star)


coor_x=[x.x for x in stars]
coor_y=[y.y for y in stars]
print(stars[6].name)
```

```python
def buscarEstrella(name):
    for star in stars:
        
        if star.name:
            # print(f'{name}: {star.name}')
            if name in star.name:
                return star
    return None

def leerConstelacion(ruta, ax, colores='#00ffff'):
    with open(ruta, 'r') as archivo:
        # Leer cada línea del archivo y hacer split por espacio
        for linea in archivo:
            datos = linea.split(",")
            datos[1] = datos[1].strip()
            datos[0]= datos[0].replace(" ", "")
            datos[1]= datos[1].replace(" ", "")
            st1 = buscarEstrella(datos[0])
            st2 = buscarEstrella(datos[1])
            
            if st1!=None and st2!=None :
                x= [st1.x, st2.x]
                y=[st1.y, st2.y]
                ax.plot(x, y, color=colores)
            elif st1==None:
                print("No existe estrella llamada: ",datos[0])
            elif st2==None:
                print("No existe estrella llamada: ",datos[1])
                
```

```python
def mostrarEstrellas(op=0):
    fig, ax = plt.subplots()

    # Establecer el color de fondo
    fig.set_facecolor('#1e1e1e')
    ax.set_facecolor('#1e1e1e')
    if op==1: #Estrellas y Boyero
         leerConstelacion('constelaciones\Boyero.txt', ax)

    if op==2: #Estrellas y Casiopea
        leerConstelacion('constelaciones\Casiopea.txt', ax)
    if op==3: #Estrellas y Cazo
        leerConstelacion('constelaciones\Cazo.txt', ax)
    if op==4: #Cygnet
        leerConstelacion('constelaciones\Cygnet.txt', ax)
    if op==5: #Geminis
        leerConstelacion('constelaciones\Geminis.txt', ax)
    if op==6: #Hydra
        leerConstelacion('constelaciones\Hydra.txt', ax)
    if op==7: #Osa Mayor
        leerConstelacion('constelaciones\OsaMayor.txt', ax)
    if op==8: #Osa Menor
        leerConstelacion('constelaciones\OsaMenor.txt', ax)
    if op==9: #TODAS LAS CONSTELACIONES
        leerConstelacion('constelaciones\Boyero.txt', ax, '#ff00ff')
        leerConstelacion('constelaciones\Casiopea.txt', ax, '#ff7514')
        leerConstelacion('constelaciones\Cygnet.txt', ax, '#0000ff')
        leerConstelacion('constelaciones\Geminis.txt', ax, '#ffff00')
        leerConstelacion('constelaciones\Hydra.txt', ax, '#00ff00')
        leerConstelacion('constelaciones\OsaMayor.txt', ax,'#00ffff')
        leerConstelacion('constelaciones\OsaMenor.txt', ax, '#c939ff')
        leerConstelacion('constelaciones\Cazo.txt', ax, 'red')
        

    ax.scatter(coor_x,coor_y, color='#ffffff', s=1)
    plt.xlim([-1, 1])
    plt.ylim([-1,1]) #plt.show()
    plt.axis('off')
    plt.savefig("todas.jpg", dpi=600)
```
